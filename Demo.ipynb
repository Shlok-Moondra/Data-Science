# Properties of array

```python
properties = np.array([[1,2,3],[5,6,7]])
print("size",np.size(properties))
print("shape",np.shape(properties))
print("Dimension",np.ndim(properties))
print("Byte",np.byte(properties))
print("DType",properties.dtype)



```

```python
arr = np.arange(12)
print("Original Array is: ",arr)

reshaped = arr.reshape((3,4))
print("Reshaped array is: ",reshaped)

## This return copy
flatten = reshaped.flatten()
print("Flatten array: ",flatten)

## This return view by changing the acutal array
raveled = flatten.ravel()
print("Raveled array: ",raveled)

print(arr)

## Transpose

transpose  = reshaped.T
print("Transpose: ",transpose)
```
